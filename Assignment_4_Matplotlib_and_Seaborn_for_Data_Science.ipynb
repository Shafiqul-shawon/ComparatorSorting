```python
gym = pd.read_csv('/content/drive/MyDrive/dataset_4assignment/gym_membership.csv')
```

# **Gym Data Exploration**

```python
# Gym data

# Display last 5 rows
print(gym.tail())

# Check data types
print(gym.dtypes)
```

```python
# Basic gym plots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Age distribution

axes[0,0].hist(gym['Age'], bins=10, color='skyblue', edgecolor='black')
axes[0,0].set_title('Age Distribution')
axes[0,0].set_xlabel('Age')
axes[0,0].set_ylabel('Frequency')


# Gender count

axes[0,1].pie(gym['gender'].value_counts(), labels=gym['gender'].unique(), autopct='%1.2f%%', colors=['pink', 'lightgreen'])
axes[0,1].set_title('Gender Count')


# Visits per week

axes[1,0].hist(gym['visit_per_week'], bins=10, color='skyblue', edgecolor='black')
axes[1,0].set_title('Visits per Week')
axes[1,0].set_xlabel('Visits per Week')
axes[1,0].set_ylabel('count of Visits per Week')


# Personal training

axes[1,1].pie(gym['personal_training'].value_counts(), labels=gym['personal_training'].unique(), autopct='%1.2f%%', colors=['pink', 'lightgreen'])
axes[1,1].set_title('Personal Training')

plt.tight_layout()
plt.show()
```

# 1. Bar chart of gym membership types

```python
# Task 1: Basic Matplotlib (3 plots)
# 1. Bar chart of gym membership types
# Use: gym['abonoment_type'].value_counts()
abonoment_type_counts = gym['abonoment_type'].value_counts()
plt.bar(abonoment_type_counts.index, abonoment_type_counts.values)
plt.title('Bar chart of gym membership types')
plt.xlabel('Gym membership types----->')
plt.ylabel('count of types----->')
plt.grid(True, which="both", linestyle="--", linewidth=0.8, alpha=0.9)
plt.savefig("/content/drive/MyDrive/dataset_4assignment/bar_abonoment_type.png",dpi=300, bbox_inches='tight')
plt.show()
```

# 2. Histogram of time in gym

```python
# 2. Histogram of time in gym
# Use: plt.hist(gym['avg_time_in_gym'])
plt.hist(gym['avg_time_in_gym'], bins=10, color='blue', edgecolor='white')
plt.title('Histogram of time in gym')
plt.xlabel('Time in gym')
plt.ylabel('Frequency')
```

#  3. Scatter plot: Age vs visits per week

```python
# 3. Scatter plot: Age vs visits per week
# Use: plt.scatter(gym['Age'], gym['visit_per_week'])

plt.scatter(gym['Age'], gym['visit_per_week'])
plt.title('Scatter plot: Age vs visits per week')
plt.xlabel('Age')
plt.ylabel('visits per week')
```

# 1. Box plot: Gender vs Age

```python
# 1. Box plot: Gender vs Age
# Use: sns.boxplot(data=gym, x='gender', y='Age')
sns.boxplot(data=gym, x='gender', y='Age')
plt.title('Box plot: Gender vs Age')
plt.xlabel('Gender')
plt.ylabel('Age')
```

# 2. Count plot: Days per week

```python
# 2. Count plot: Days per week
# Use: sns.countplot(data=gym, x='days_per_week')
plt.figure(figsize=(20,4))
sns.countplot(data=gym, x='days_per_week')
plt.xticks(rotation=90)
plt.title('Count plot: Days per week')
plt.xlabel('Days per week')
plt.ylabel('count of Days per week')
```

# 3. Bar plot: Average visits by membership type

```python
# 3. Bar plot: Average visits by membership type
# Use: sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')
sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')
plt.title('Bar plot: Average visits by membership type')
plt.xlabel('Membership type')
plt.ylabel('Average visits')
```
